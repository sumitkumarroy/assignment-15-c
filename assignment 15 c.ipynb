```python
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
iris = sns.load_dataset('iris')
iris.head()
```

```python
#Q1
# scatter plot
sns.scatterplot(x = iris.sepal_length, y =iris.petal_width)

#Barplot
sns.barplot(x = iris.sepal_length, y =iris.sepal_width)

#catgorical plot
sns.catplot(x = iris.species, y =iris.sepal_width)

#boxplot
sns.boxplot(x=iris.species, y=iris.petal_length)

#pairplot
sns.pairplot(iris)
```

```python
#Q2
fmri = sns.load_dataset('fmri')
fmri.columns

sns.lineplot(x = fmri.timepoint,y = fmri.signal)
```

```python
#Q3
titanic = sns.load_dataset('titanic')
titanic.columns
print(titanic)

sns.boxplot(x = titanic.pclass,y =titanic.age)
```

```python
#Q4
diamonds = sns.load_dataset('diamonds')
diamonds.columns
print(diamonds)

sns.histplot(data = diamonds, x='price', hue = "cut", multiple='stack')
```
